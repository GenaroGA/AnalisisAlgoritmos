## 12 Factorial

Estos algoritmos buscan el elemento $x$ en un arreglo $A$ en el intervalo $[r,l]$, regresando su indice si esta y $0$ en caso contrario.

### 12.1.1 Algoritmo Iterativo

```julia
function Search_I(A::Vector{Int64}, x::Int64, r::Int64, l::Int64)::Int64
    for i::Int64 in r:l
        if A[i] == x
            return i
        end
    end
    return 0 
end
```

### 12.1.2 Algoritmo Recursivo

```julia
function Search_R(A::Vector{Int64}, x::Int64, r::Int64, l::Int64)::Int64
    if A[r] == x
        return r
    elseif r == l
        return 0
    else
        return Search_R(A, x, r + 1, l)
    end
end
```

### 12.2.1 Verificación (Iterativo)

Vemos que el for-loop de algoritmo unicamente revisa cada elemento de $A$ en el intervalo $[r,l]$, no modifica $A$ en ningun momento, regresa $i$ si y solo si $A[i] = x$, en otro caso regresa 0. Por lo tanto el algoritmo es correcto.

### 12.2.2 Verificación (Recursivo)

Vemos que el algoritmo termina solo en dos casos, si $A[r] = x$ o si $r = l$. Si $A[r] = x$ se regresa $r$. Si $r = l$ entonces ya se revisaron todos las posibles posiciones de $x$ sin encontrarlo, gracias a la linea 6, y regresa $0$. Ambos casos son correctos por lo tanto el algoritmo es correcto.

### 12.3.1 Análisis del tiempo de ejecución (Iterativo)

Definamos $n = l - r + 1$. Entonces el tiempo de ejecucion del algoritmo es $O(n)$, ya que el for-loop de las lineas 1-5 se ejecuta a lo mas $n + 1$ veces.

### 12.3.2 Análisis del tiempo de ejecución (Recursivo)

Vemos que el algoritmo termina cuando se encuentra $x$ o cuando $r = l$. Por la linea 6 vemos que cada llamada recursiva $r$ aumenta en uno. Por lo tanto el tiempo de ejecucion igualmente sera $O(n)$.
