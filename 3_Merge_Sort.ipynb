<h2>Merge Sort</h2>

Merge Sort es un algortimo de ordenación que opcupa el enfoque "Divide y Vecerás", este algoritmo se divide en 2 funciones para compleatr el proceso:
<br>
<b>1- Merge:</b> Se encarga de combinar arreglos que previemente fueron ordeandos y devolver un nuevo arreglo que siga siendo ordenado. Justo aquí es cuando se usa el concepto de dividir y vencer ya que es más fácil combinar odenadamente 2 arreglos que previemanete fueron odenados individualmente.
<br>
<b>2- Merge Sort:</b> Se encarga de usar la herramienta de recursión para ocupar la función Merge y poder pasarle los parámetros necesarios. La manera en que lo hace es divididiendo el arreglo en mitades hasta llegar al caso base y posterioirmente usar la función Merge para combinar "los casos base" en un arreglo nuevo y ordenado que conforme avance la recursión será ada vez mas grande para al final tener el arreglo original ordenado.


<h4>Merge</h4>
El pseudo algorimto que vamos a usar como referencia es el siguiente (Cormen, p. 31):
<br>
<img src="https://github.com/GenaroGA/AnalisisAlgoritmos/blob/main/img/pseudo_merge.JPG?raw=true" alt="Pseudocodigo Merge" />

```python
def Merge(A, p, q, r):
    
    #q es un parámetro que indica el elemnto en la mitad del arreglo y se calcula en la función Merge Sort
    #p es el primer elemento del arreglo
    #r es el último elemento del arreglo
    n1 = q - p + 1
    #n1 será la longitud de la primera mitad del arreglo
    n2 = r - q
    #n2 será la longitud de la segunda mitad del arreglo
 
    #declaramos los arreglos temporales L (left/izquierda) R (right/derecha)
    L = [] 
    R = [] 
 
    #En los ciclos for vamos a copiar la información del arreglo a los dos arreglos nuevos
    #Usando las logitudes como n1 y n2 rango del ciclo
    #En el arreglo L se almacena la primera mitad y en R la segunda mitad
    for i in range(0, n1):
        L.append(A[p + i])
 
    for j in range(0, n2):
        R.append(A[q + 1 + j])
        
    #agregamos los centinelas para las comparaciones futuras
    L.append(1000000)
    R.append(1000000) 
    
    """"Si se hacen estas impresiones se pude observar como trabaja el algoritmo a detalle
    print("L:")
    print(L)
    print("R:")
    print(R)"""
    #inicializamos las variables en 0
    i = 0 
    j = 0 

    for k in range (p, r+1):
        #comparamos los elementos de cada arreglo y agregamos el menor a el arreglo ordenado
        #del arreglo donde salga el menor aumentamos el índice para que en la siguiente comparación sea el elemnto posterior al anterior
        if L[i] <= R[j]:
            A[k] = L[i]
            i = i + 1
        else:
            A[k] = R[j]
            j = j + 1

    return A

 
```

<h4>Merge Sort</h4>
El pseudo algorimto que vamos a usar como referencia es el siguiente (Cormen, p. 34):
<br>
<img src="https://github.com/GenaroGA/AnalisisAlgoritmos/blob/main/img/pseudo_merge_sort.JPG?raw=true" alt="Pseudocodigo Merge Sort" />

```python
import math
def MergeSort(A, p, r):
    if p < r:
        #usamos la función floor para calcular el elemento que se encuentra en la posición media
        q = math.floor(p+(r-p)/2)
 
        #recursión para la primera mitad
        MergeSort(A, p, q)
        #recursión para la segunda mitad
        MergeSort(A, q+1, r)
        #combinar los arreglos ya ordenado con merge
        Merge(A, p, q, r)
    return(A)
 
```

<h3>Prueba del algortimo</h3>
Se va a correr el algoritmo para verficiar que funcione correctamente

```python
A = [4,1,2,0,-5]
p = 0
r = 4

x = MergeSort(A, p, r)
print(x)
```

<h3>Verificación del algoritmo</h3>
<b>Merge Sort (Recursivo):</b><br>
<b>Caso base:</b> Cuando el arreglo que vamos a ordenar tiene 2 elementos, el cual se ordena mediante la función Merge sin nignún problema. Si el arreglo tuviera un sólo elemento si podría causar un mal funcionamiento de la función Merge.<br><br>
Notemos que el algortimo se enceuntra completamente contenido dentro de un if con la condición p < r, de esta manera como se toma en cuenta un caso base, siempre tiene que haber mínimo 2 elementos en el arreglo en caso de que tenga menos ya no entra al if.
Después, en la línea 3 se está ejecutando el mismo procedimeinto para la mitad de elementos que va de p hasta q (primera mitad).<br><br>
En la línea 4 se ejecuta el merge sort pero igualmente para la mitad de elementos ya que va de q+1 hasta r (segunda mitad).<br><br>
Por lo tanto cada vez que se ejecuta la recursión esta disminuyendo a la mitad la cantidad de elementos en el arreglo por lo que forzosamente va a reducirse hasta llegar al caso base y ya no entrar en el if<br><br>
<b>Orden:</b> El algortimo funciona ya que la función Merge combina ordenadamente 2 arreglos ordenados.<br>

Como consecuencia de cómo está hecha la recursión, la primera vez que se ejecute la función Merga será en el caso base, cuando 2 el arreglo L y el arreglo R tengan 1 elemento respectivamente, por lo que si solo tienen un elemento... están ordenados y la función Merge los va a cominar ordenadamente, posteriormente, esos arreglos ordandos serán los que volverán a pasar como parámetros por lo que se garantiza que la función Merge siempre va a trabajar con dos arrgelos orenados y por lo tanto siempre devolverá un arreglo ordenado.




<h3>Timepo de ejecución</h3>
<b>Función Merge:</b> &Theta; (n)<br>
Para poder confirmar que el timepo de ejecución es &Theta; (n), basta con observar que el ciclo for de la línea 4-5 toma  &Theta; (n_1) y el ciclo for de la línea 6-7 toma &Theta; (n_2), por lo que de la línea 4-7 toma un timepo de  &Theta; (n_1 + n_2) =  &Theta; (n), mientras que el ciclo for de la línea 12-17 tiene n iteraciones y todas las demás líneas de la función tienen un tiempo constante.
